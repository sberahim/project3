```python
#import dependency
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load covid csv file
Cleaned_covid_load = "cleaned_covid.csv"
#create panda data frame
Covid_data = pd.read_csv(Cleaned_covid_load)
#print out covid data
print(Covid_data)
```

```python
# Replace null values with "0"
Covid_new_data = Covid_data.fillna(0)

# Print the resulting DataFrame
print(Covid_new_data)
```

```python
# Filter out non-country locations
country_data = Covid_new_data[Covid_new_data['location'].str.contains('[A-Za-z]+')]

# Get the names of unique attributes under location
unique_attributes = country_data['location'].unique()

print("Names of attributes under 'location':")
for attribute in unique_attributes:
    print(attribute)
```

```python
#TOTAL COVID CASESS OF THE TOP 20 COUNTRIES

# Filter out non-country locations
country_data = Covid_new_data[Covid_new_data['location'].str.contains('[A-Za-z]+')]
country_data = country_data[~country_data['location'].isin(['Asia', 'World', 'High income', 'North America', 'Middle income','Upper middle income',  'Lower middle income', 'Europe', 'European Union'])]

# Group the data by location and calculate the total cases
grouped_data = country_data.groupby('location')['total_cases'].sum().reset_index()

# Sort the data in descending order based on total_cases
sorted_data = grouped_data.sort_values('total_cases', ascending=False)

# Set up the figure and axis
fig, ax = plt.subplots(figsize=(10, 20))

# Filter out top 10 countries
top_countries = sorted_data.head(20)

# Create the pie chart
ax.pie(top_countries['total_cases'], labels=top_countries['location'], autopct='%1.1f%%')

# Customize the chart
ax.set_title('Percentage of Total COVID-19 Cases by Country (Top 20)')

# Display the chart
plt.show()
```

```python
# Convert the "date" column to datetime object
Covid_new_data['date'] = pd.to_datetime(Covid_new_data['date'], format='%d/%m/%Y')

# Filter the dataframe for the relevant years
years = [2020]
filtered_data = Covid_new_data[Covid_new_data['date'].dt.year.isin(years)]

# Group the data by month and calculate the total cases
monthly_cases = filtered_data.groupby(filtered_data['date'].dt.to_period('M'))['total_cases'].sum()

# Convert the monthly periods to strings
x_values = monthly_cases.index.strftime('%b %Y')

# Create the line graph
plt.plot(x_values, monthly_cases.values)
plt.xlabel('Month')
plt.ylabel('Total Cases')

# Get the year for the heading
year = filtered_data['date'].dt.year.unique()[0]

# Set the heading
plt.title("Monthly Total Cases Year 2020")

plt.xticks(rotation=45)
plt.show()
```

```python
# Filter the dataframe for the relevant years
years = [2020, 2021, 2022, 2023]
filtered_data = Covid_new_data[Covid_new_data['date'].dt.year.isin(years)]

# Create separate dataframes for each year
data_2020 = filtered_data[filtered_data['date'].dt.year == 2020]
data_2021 = filtered_data[filtered_data['date'].dt.year == 2021]
data_2022 = filtered_data[filtered_data['date'].dt.year == 2022]
data_2023 = filtered_data[filtered_data['date'].dt.year == 2023]

# Plot 'total_cases' and 'total_deaths' for each year in different colors
plt.scatter(data_2020['total_cases'], data_2020['total_deaths'], c='blue', label='2020')
plt.scatter(data_2021['total_cases'], data_2021['total_deaths'], c='green', label='2021')
plt.scatter(data_2022['total_cases'], data_2022['total_deaths'], c='orange', label='2022')
plt.scatter(data_2023['total_cases'], data_2023['total_deaths'], c='red', label='2023')

plt.xlabel('Total Cases')
plt.ylabel('Total Deaths')
plt.title('Scatter Plot of Total Cases vs Total Deaths')
plt.legend()

plt.show()
```
