```python
# Dependencies and Setup
import pandas as pd

# csv files
country_codes = "data/country_codes_and_coordinates.csv"
covid = "data/owid-covid-data.csv"

face = "data/face-covering-policies-covid.csv"
transport = "data/public-transport-covid.csv"
home = "data/stay-at-home-covid.csv"

# Read csv files
country_codes_data = pd.read_csv(country_codes)
covid_data = pd.read_csv(covid)

face_data = pd.read_csv(face)
transport_data = pd.read_csv(transport)
home_data = pd.read_csv(home)

country_codes_data
```

```python
country_codes_data = country_codes_data.rename(columns={
    "Country":"location",
    "Alpha-3 code":"iso_code"})

country_codes_data = country_codes_data.drop(columns=['location'])
country_codes_data
```

```python
covid_data
```

```python
merged_df = pd.merge(covid_data, country_codes_data, how="left", on=["iso_code"])
```

```python
merged_df
```

```python
face_data = face_data.rename(columns={
    "Code":"iso_code",
    "Day":"date"})

face_data = face_data.drop(columns=['Entity'])

face_data
```

```python
transport_data = transport_data.rename(columns={
    "Code":"iso_code",
    "Day":"date"})

transport_data = transport_data.drop(columns=['Entity'])

transport_data
```

```python
home_data = home_data.rename(columns={
    "Code":"iso_code",
    "Day":"date"})

home_data = home_data.drop(columns=['Entity'])

home_data
```

```python
merged_df2 = pd.merge(merged_df, face_data, how="inner", on=["iso_code", "date"])
```

```python
merged_df2
```

```python
merged_df3 = pd.merge(merged_df2, transport_data, how="inner", on=["iso_code", "date"])

merged_df3
```

```python
merged_df4 = pd.merge(merged_df3, home_data, how="inner", on=["iso_code", "date"])

merged_df4
```

```python
cleaned_df = merged_df4[["iso_code", "continent", "location", "date", "total_cases", "total_deaths", "total_vaccinations", "total_vaccinations_per_hundred", 
                        "total_tests", "total_tests_per_thousand", "population", "population_density", "Latitude (average)", "Longitude (average)", "facial_coverings", "close_public_transport", "stay_home_requirements" ]]

cleaned_df = cleaned_df.rename(columns={
    "Latitude (average)": "Latitude",
    "Longitude (average)": "Longitude"})

cleaned_df
```

```python
cleaned_df.to_csv("output/cleaned_covid.csv", index=False, header=True)
```
