```python
# Dependencies and Setup
import pandas as pd

# csv files
country_codes = "data/country_codes_and_coordinates.csv"
covid = "data/owid-covid-data.csv"

# Read csv files
country_codes_data = pd.read_csv(country_codes)
covid_data = pd.read_csv(covid)

country_codes_data
```

```python
country_codes_data = country_codes_data.rename(columns={
    "Country":"location",
    "Alpha-3 code":"iso_code"})

country_codes_data
```

```python
covid_data
```

```python
merged_df = pd.merge(covid_data, country_codes_data, how="left", on=["location", "iso_code"])
```

```python
merged_df
```

```python
merged_df.columns
```

```python
cleaned_df = merged_df[["iso_code", "continent", "location", "date", "total_cases", "total_deaths", "total_vaccinations", "total_vaccinations_per_hundred", 
                        "total_tests", "total_tests_per_thousand", "population", "population_density", "Latitude (average)", "Longitude (average)" ]]

cleaned_df = cleaned_df.rename(columns={
    "Latitude (average)": "Latitude",
    "Longitude (average)": "Longitude"})

cleaned_df
```
