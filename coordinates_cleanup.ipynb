```python
import pandas as pd
import os
import csv
```

```python
rawcoords_csvpath = os.path.join("data","countries_raw.csv")
```

```python
countries_df = pd.read_csv(rawcoords_csvpath)
countries_df.head()
```

```python
for i, col in enumerate(countries_df.columns):
    countries_df.iloc[:, i] = countries_df.iloc[:, i].str.replace('"', '')
countries_df.head()
```

```python
coords_df = countries_df[["Country", "Alpha-3 code", "Latitude (average)", "Longitude (average)"]]
coords_df.head()
```

```python
coords_df.to_csv("data/country_codes_and_coordinates.csv", index=False, header=True)
```
